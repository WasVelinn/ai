# 1. 자연어 처리
- 일상적인 언어를 처리
- 자연어 처리 분야
    * 자연어 이해 : 형태소분석 -> pos tagging -> 의미 분석
    * 자연어 생성 : 대화분석 -> 다음 문장이나 단어를 출력(RNN, LSTM, ... , GAN, 생성형 알고리즘)

# 3.1 HanNanum

```python
from konlpy.tag import Hannanum
hannanum = Hannanum(jvmpath=None,
                    max_heap_size=1024) # 기본값:1024, '1g', '2048m'
hannanum.analyze(text)
```

- pos tagging chart 다운로드
    * 구글에서 konlpy docs검색 후 konlpy.org
    * api 메뉴 클릭 -> "품사 태깅 클래스 간 비교(Comparision between 

```python
# 형태소분석 : morphs
print(hannanum.morphs(text))
```

```python
# 명사만 추출
print(hannanum.nouns(text))
```

```python
# 품사태그
print(hannanum.pos(text)) # 기본값 : ntags=9
```

```python
# 퀴즈1 : text 중에서 형용사만 추출
pa_list = []
for wordtag in hannanum.pos(text, ntags=22):
    if wordtag[1] == 'PA':
        pa_list.append(wordtag)
print(pa_list)
```

```python
# 퀴즈2 : text 중에서 보통명사만 추출
nc_list = []
for wordtag in hannanum.pos(text, ntags=22):
    if wordtag[1] == 'NC':
        nc_list.append(wordtag)
print(nc_list)
```

## 5. 워드 클라우드
- pip install wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import nltk
from nltk.tokenize import RegexpTokenizer # 단어 추출(정규표현식을 활용한)
from nltk.tag import pos_tag
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
# 단어추출
ret = RegexpTokenizer('[\w]{3,}')
words = ret.tokenize(emma)
# 명사만 추출
emma_tags = pos_tag(words)
# noun_list = [emma_tag[0] for emma_tag in emma_tags if emma_tag[1] in ('NN', 'NNS')]
noun_list = [word for word, tag in emma_tags if tag in ('NN', 'NNS')]
emma_noun = ' '.join(noun_list)
```

```python
emma_noun
```

```python
# 명사 갯수
len(set(noun_list))
```

```python
wordc = WordCloud()
wordc.generate(emma_noun)
plt.figure(figsize=(10,5))
plt.imshow(wordc)
plt.show()
```

```python
wordc = WordCloud(background_color='white', # 배경색
                  max_words=400, # 표시될 단어의 최대 갯수
                  relative_scaling = 0.2, # 단어 빈도에 따른 크기 차이 조정
                  colormap = 'viridis' # 컬러맵
)
wordc.generate(emma_noun)
plt.figure(figsize=(10,5))
plt.imshow(wordc)
plt.show()
```

```python
# data 명사 추출 후 ' '.join(명사추출리스트)
from konlpy.tag import Kkma
kkma = Kkma()
# word_list = kkma.nouns(data)
# 불용어 처리를 여기서 해도 됨
word_list = [word for word, tag in kkma.pos(data) if tag in ['NNG', 'NNP']]
data_noun = ' '.join(word_list)
```

```python
wordc = WordCloud(background_color='white', # 배경색
                  max_words=400, # 표시될 단어의 최대 갯수
                  relative_scaling = 0.2, # 단어 빈도에 따른 크기 차이 조정
                  colormap = 'viridis', # 컬러맵
                  font_path= 'HanSantteutDotum-Bold.ttf' # ttf 폰트 설정
)
wordc.generate(data_noun)
plt.figure(figsize=(10,5))
plt.imshow(wordc)
plt.show()
```

```python
# 불용어 처리(불용어 지정 + 불용어사전)
from wordcloud import STOPWORDS
불용어 = STOPWORDS | {'대통령', '때'}
불용어 = {'대통령', '때'}
불용어 = set(['대통령', '때'])
```

```python
wordcloud = WordCloud(background_color='white',
                      max_words=300,
                      font_path = 'HanSantteutDotum-Bold.ttf',
                      relative_scaling=0.2,
                      colormap='viridis',
                      stopwords=불용어)
wordcloud.generate(data_noun)
plt.figure(figsize=(10,4))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# 마스킹 : 워드 클라우드를 지정된 마스크 이미지에 맞도록 설정
from PIL import Image
import numpy as np
img = Image.open('data/test.png')
mask = np.array(img)
# plt.imshow(img)
# mask.shape
wordcloud = WordCloud(background_color='white',
                      max_words=300,
                      font_path = 'HanSantteutDotum-Bold.ttf',
                      relative_scaling=0.2,
                      colormap='viridis',
                      stopwords=불용어,
                      mask=mask)
wordcloud.generate(data_noun)
plt.figure(figsize=(10,4))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# 단어의 빈도수 시각화(nltk.Text)
# 워드클라우드에서는 data_noun(출현한 명사들을 ' '로 연결한 문자)
# nltk.Text().plot(출현한 명사들 list) : word_list ['대한민국', '헌번', '유구', ...]
import nltk
import matplotlib.pyplot as plt
from konlpy.tag import Komoran
from konlpy.corpus import kolaw
data = kolaw.open('constitution.txt').read()
komoran = Komoran()
# 일반명사(NNG), 고유명사(NNP), 의존명사(NNB)
word_list = [token for token, tag in komoran.pos(data) if tag in ('NNG', 'NNP', 'NNB')]
data_text = nltk.Text(word_list)
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,4))
data_text.plot(15)
plt.show()
```

```python
data_text.vocab().most_common(15)
```

```python
list(zip(*data_text.vocab().most_common()))
```

```python
words, counts = zip(*data_text.vocab().most_common(15))
plt.figure(figsize=(15,4))
plt.plot(words, counts)
plt.xticks(rotation=10) # rotation 기본값 : 0
plt.xlabel('Samples')
plt.ylabel('Counts')
plt.grid()
plt.show()
```
