```python
# 1. 임신횟수
# 2. 2시간내 포도당 농도
# 3. 혈압
# 4. 피부접힘 두께
# 5. 인슐린수치
# 6. BMI체질량지수
# 7. 가족력
# 8. 나이
# 9. 당뇨병발병여부
```

```python
# csv파일을 numpy 배열로 읽어오기, np.loadtxt(), np.genfromtxt()
dataset = np.loadtxt('data/pima-indians-diabetes.csv', delimiter = ',')
dataset.shape
```

```python
# 학습데이터셋(학습용 데이터)
X_train = dataset[:700, :-1]
y_train = dataset[:700, -1]

# 테스트데이터셋(시험용 데이터:test dataset)
X_test = dataset[700:, :-1]
y_test = dataset[700:, -1]

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## 2. 모델 구성

```python
model = Sequential()
model.add(Input(shape=(8,)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
```

## 학습과정 설정

```python
# model.compile(loss='binary_crossentrophy', optimizer='adam', metrics=['accuracy'])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
```

## 4. 학습

```python
hist = model.fit(X_train, y_train, # 훈련데이터
                 epochs=350,
#                batch_size=350,
#                validation_split=0.2 # 검증 데이터 비율(훈련데이터셋 중 20%를 검증용 데이터로)
                 
                 )
```

## 5. 모델 평가하기(학습과정보기, 평가)

```python
hist.history.keys()
```

```python
# 학습과정 표시하기(validation dataset 없는 경우)
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'], 'r', label='train loss')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax = loss_ax.twinx()
acc_ax.plot(hist.history['binary_accuracy'], 'b', label='train accuracy')
acc_ax.set_ylabel('accuracy')
loss_ax.legend(loc='center right')
acc_ax.legend(loc='lower right')
acc_ax.legend(bbox_to_anchor=(0.999, 0.65))
plt.show()
```

```python
# 모델 평가(X)
model.evaluate(X_train, y_train)
```

```python
# 모델 평가(O)
score = model.evaluate(X_test, y_test)
print("loss:{}, 정확도 : {}%".format(score[0], score[1]*100))
```

## 7. 교차표
- 실제값과 예측값

```python
y_hat = (model.predict(X_test)>0.5).astype(float)
cnt1 = 0; cnt2 = 0; cnt3 = 0; cnt4 = 0
# y_hat.shape, y_test.shape
for h, y in zip(y_hat.reshape(-1), y_test.reshape(-1)):
    if h == 1.0 and y == 0.0 :
        cnt1 += 1
    elif y == 1.0 and h == 0.0 :
        cnt2 += 1
    elif y == 0.0 and h == 1.0 :
        cnt3 += 1
    else:
        cnt4 += 1
print(cnt1, cnt2)
print(cnt3, cnt4)
```

```python
ctab = pd.crosstab(y_test, y_hat.reshape(-1))
ctab.index.name = '실제값'
ctab.columns.name = '예측값'
ctab
```

```python
# - 1. 데이터셋 생성 및 전처리 : 훈련셋600, 검증셋100, 테스트셋68으로 분리, 종속변수의 원핫인코딩
dataset = np.loadtxt('data/pima-indians-diabetes.csv', delimiter = ',')
dataset.shape
```

```python
X_train = dataset[:600, :-1]
y_train = dataset[:600, -1]

X_test = dataset[700:, :-1]
y_test = dataset[700:, -1]

X_val = dataset[600:700, :-1]
y_val = dataset[600:700, -1]
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
```

```python
le = LabelEncoder()
label_y_train = le.fit_transform(y_train)
onehot_y_train = to_categorical(label_y_train)
```

```python
label_y_test = le.fit_transform(y_test)
onehot_y_test = to_categorical(label_y_test)
```

```python
label_y_val = le.fit_transform(y_val)
onehot_y_val = to_categorical(label_y_val)
```

```python
model = Sequential()
model.add(Input(shape=(8,)))
model.add(Dense(units=12, activation='relu'))
model.add(Dense(units=8, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model.fit(X_train, onehot_y_train, # 훈련데이터
                 epochs=350,
#                batch_size=350,
                 validation_data=[X_val, onehot_y_val] # 검증 데이터 비율(훈련데이터셋 중 20%를 검증용 데이터로)
                 
                 )
```

```python
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'], 'r', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'g', label='train loss')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax = loss_ax.twinx()
acc_ax.plot(hist.history['accuracy'], 'b', label='train accuracy')
acc_ax.set_ylabel('accuracy')
loss_ax.legend(loc='center right')
acc_ax.legend(loc='lower right')
acc_ax.legend(bbox_to_anchor=(0.999, 0.65))
plt.show()
```

```python
# 모델 평가하기(test dataset으로)
score = model.evaluate(X_test, onehot_y_test)
print('Test Dataset loss : {}, accuracy(정확도) : {}'.format(score[0], score[1]))
```

```python
score # loss와 compile에서 metrics = ['accuracy']지정으로 accuracy가 저장
```

```python
# 교차표(원핫인코딩을 이용한 분류분석, 이진분류)
# pd.crosstab(실제값을1차원, 예측값을1차원)
Y_test = all(onehot_y_test.argmax(axis=1) == y_test)
```
