## 2. 사이킷런(sklearn) 이용
- 원핫인코딩을 하지 않고 라벨인코딩까지만 해야 작동. 원핫인코딩은 작동되지 않음.

```python
from sklearn import datasets # iris데이터
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
# 1. 데이터
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target
print(iris_X[::50], iris_y[::50])
train_X, test_X, train_y, test_y = train_test_split(iris_X, iris_y,
                                                    test_size = 0.1, # train:test의 기본값은 0.75 : 0.25
                                                    random_state=4,
                                                    stratify = iris_y)
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

```python
# 세이킷런에서 제공하는 다층퍼셉트론 기반의 분류 알고리즘 api
mlp = MLPClassifier(hidden_layer_sizes=(50,30),
                    activation='relu', # tanh
                    solver = 'adam', # optimizer 설정값
                    max_iter = 1000, # 최대학습횟수
                    early_stopping= True, # 조기 종료 활성화
                    n_iter_no_change=150, # 검증셋 손실함수가 n회동안 개선되지 않으면 조기종료
                    validation_fraction=0.1 # 검증셋 비율
                    )
```

```python
mlp.fit(train_X, train_y)
```

```python
mlp.score(test_X, test_y)
```

```python
mlp.predict(test_X[0:1])
```

## 클래스생성

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
```

```python
class DNNClassfier:
    @staticmethod
    def build(input_dim=4, activation='relu', output_dim=3):
#         model = Sequential()
#         model.add(Input(input_dim,))
#         model.add(Dense(50, activation=activation))
#         model.add(Dense(30, activation=activation))
#         model.add(Dense(3, activation=activation))
        model = Sequential([
            Input(input_dim,),
            Dense(50, activation=activation),
            Dense(30, activation=activation),
            Dense(output_dim, activation='softmax'),
        ])
        #학습과정 설정
        model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
        return model
```

```python
# 1. 데이터 생성 및 전처리
iris_Y = pd.get_dummies(iris_y).to_numpy()
# train_X, test_X, train_y, test_y = train_test_split(iris_X, iris_Y,
#                                                     test_size = 0.1, # train:test의 기본값은 0.75 : 0.25
#                                                     random_state=4,
#                                                     stratify = iris_y)
# print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
train_Y = pd.get_dummies(train_y).values
test_Y = pd.get_dummies(test_y).values
# 2. 모델
model = DNNClassfier.build(4, 'relu', 3)
# 3. 모델 학습
earlystopping = EarlyStopping(patience=150)
hist = model.fit(train_X, train_Y, epochs=1000, validation_split=0.1, callbacks=[earlystopping])
```

# 4. 함수 api 이용 : 병렬처리

```python
from tensorflow.keras.layers import concatenate
```

```python
train_X.shape, train_Y.shape, test_X.shape, test_Y.shape
```

```python
# 모델
input_ = Input(shape=(4,))
dense1 = Dense(50, activation='relu')(input_)
dense2 = Dense(80, activation='relu')(input_)
dense3 = Dense(30, activation='relu')(input_)
x = concatenate([dense1, dense2, dense3])
# x = Dense(160, activation='relu')(input_)의 병렬처리 구현
output = Dense(units=3, activation='softmax')(x)
model = Model(inputs=input_, outputs=output)
model.summary()
```

```python
# 학습 정의(설정), 학습
model.compile(loss='categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy']
              )
```

```python
earlystopping = EarlyStopping(patience=150)
hist = model.fit(train_X, train_Y, epochs=1000, validation_split=0.1, callbacks=[earlystopping])
```

```python
model.evaluate(test_X, test_Y)
```
