```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import matplotlib.pyplot as plt
```

```python
# 1 데이터생성 및 전처리
iris = sns.load_dataset('iris')
# 독립변수와 종속변수 분리
iris_X = iris.iloc[:, :-1].to_numpy()
iris_y = iris.iloc[:, -1]
iris_Y = pd.get_dummies(iris_y).values
# iris_Y[::50]
# 훈련셋 : 테스트셋 = 8 : 2
X_train, X_test, Y_train, Y_test = train_test_split(iris_X, iris_Y,
                                                    test_size=0.2,
                                                    stratify=iris_Y,
                                                    random_state=4)
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
pd.Series(np.argmax(Y_train, axis=1)).value_counts()
```

```python
pd.Series(np.argmax(Y_test, axis=1)).value_counts()
```

```python
# 2. 모델 구성 (입력4 - 출력3)
model = Sequential()
model.add(Dense(units=128, input_dim=4, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
# 2. 모델 구성 (입력4 - 출력3)
model = Sequential(
        Input(4,),
        Dense(units=128, activation='relu'),
        Dense(units=64, activation='relu'),
        Dense(units=32, activation='relu'),
        Dense(units=3, activation='softmax'))
```

```python
# 3. 모델학습과정
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 4. 학습 : epoch마다 val_accuracy(val_loss, accuracy, loss)값이 좋을 때 모델을 자동 저장 콜백 추가
            # 모델을 저장할 경로 : ./model/iris-100-val0.6521.h5
import os
model_save_folder = './model/'
if not os.path.exists(model_save_folder):
    os.mkdir(model_save_folder) # model_save_folder가 없으면 폴더 생성
# 조기 종료 콜백
earlystopping = EarlyStopping(patience=40) # monitor = 'val_loss' 기본값
# 모델 자동 저장 콜백
file = model_save_folder + 'iris - {epoch:03d} - val{val_accuracy:.4f}.h5'
checkpoint = ModelCheckpoint(filepath = file, # 저장경로
                             monitor = 'val_loss', # 모니터링할 지표
                             save_best_only = True, # 지표가 개선된 경우만 저장. False면 매번 저장
                             mode = 'auto', # 값이 클수록 저장. 'min'은 값이 작을수록 저장. auto도 있음
                             verbose = 1) # 저장 여부 로그 출력 (0:출력안함)
hist = model.fit(X_train, Y_train,
                 # batchsize = 50, # 한 배치사이즈마다 w,b조정
                 epochs=300,
                 validation_split = 0.2,
                 callbacks = [checkpoint, earlystopping]
                 )
```

```python
# 모델 평가
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'], 'r', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'g', label='validation loss')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax = loss_ax.twinx()
acc_ax.plot(hist.history['accuracy'], 'b', label='train accuracy')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='validation accuracy')
acc_ax.set_ylabel('accuracy')
loss_ax.legend(loc='center right')
acc_ax.legend(loc='lower right')
acc_ax.legend(bbox_to_anchor=(0.999, 0.65))
plt.show()
```

```python
# 모델 평가
model.evaluate(X_test, Y_test)
```

```python
# 교차표
real = np.argmax(Y_test, axis=1)
pred = np.argmax(model.predict(X_test), axis=1)
ctab = pd.crosstab(real, pred)
ctab.columns.name = 'predict'
ctab.columns = list(iris['species'][::50]) # ['setosa', 'versicolor', 'virginica']
ctab.index
```
