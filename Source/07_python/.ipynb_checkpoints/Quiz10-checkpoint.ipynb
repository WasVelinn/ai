```python
# 11
iris_grouped = iris.groupby('species')
l = []
m = []
for speciename, group in iris_grouped:
    l.append(group)
for idx, (speciename, group) in enumerate(iris_grouped):
    X = l[idx].iloc[:,:-1]
    avg = X.apply(np.average, axis=0)
    m.append(X.apply(lambda x : round(x-avg,2), axis=1))
n = pd.concat((m[0],m[1],m[2]))
n.take([0,1,2,50,51,52,100,101,102])
```

```python
# 11-2
mean_by_species = iris.groupby(iris.species).mean()
result = iris.apply(lambda series : series - mean_by_species.loc[series.species], axis=1)
result.groupby(iris.species).take(range(3))
```
